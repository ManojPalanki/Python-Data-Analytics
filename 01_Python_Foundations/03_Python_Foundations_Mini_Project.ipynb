# Mini Project: Business Sales KPI Summary Report

## Scenario

You are working as a **Junior Data Analyst** for a retail company. The management team wants a simple monthly business performance report.

Your task is to use **Python fundamentals** to store the given business data, calculate key performance indicators (KPIs), and display a clear and formatted summary report.

## Business Data

* Company Name: Nova Retail
* Reporting Month: June 2026
* Total Revenue: ₹850,000
* Total Cost: ₹620,000
* Total Orders: 1,700
* Total Customers: 1,250
* Previous Month Revenue: ₹780,000

## Tasks

1. Create variables to store all the given business data using appropriate data types.

2. Ask the user to enter the name of the analyst preparing the report using `input()`.

3. Calculate the following business KPIs:

   * Total Profit
   * Profit Margin Percentage
   * Average Order Value
   * Revenue Growth Percentage

4. Use the following formulas:

   * `Total Profit = Total Revenue - Total Cost`
   * `Profit Margin % = (Total Profit / Total Revenue) * 100`
   * `Average Order Value = Total Revenue / Total Orders`
   * `Revenue Growth % = ((Current Revenue - Previous Revenue) / Previous Revenue) * 100`

5. Display the final **Monthly Business KPI Report** using `print()` and f-strings.

6. Format all percentage values to **2 decimal places**.

7. Format all monetary values with **commas and 2 decimal places**.

8. Add meaningful comments to explain the important parts of your code.

9. Follow proper Python variable naming conventions.

## Expected Output Structure

```text
========================================
       MONTHLY BUSINESS KPI REPORT
========================================

Company Name        :
Reporting Month     :
Prepared By         :

Total Revenue       :
Total Cost          :
Total Profit        :
Total Orders        :
Total Customers     :

Profit Margin       :

Revenue Growth      :

========================================
```

## Bonus Challenge

Calculate **Revenue Per Customer** using the available business data and include it in the final report.

Explain in one or two sentences what this KPI tells management about business performance.


````python
# Store business data in variables
company_name = "Nova Retail"
reporting_month = "June 2026"
total_revenue = 850000
total_cost = 620000
total_orders = 1700
total_customers = 1250
previous_month_revenue = 780000

# Ask the user to enter the analyst name
analyst_name = input("Enter the analyst name: ")

# Calculate KPIs
total_profit = total_revenue - total_cost
profit_margin = (total_profit / total_revenue) * 100
average_order_value = total_revenue / total_orders
revenue_growth = ((total_revenue - previous_month_revenue) / previous_month_revenue) * 100
revenue_per_customer = total_revenue / total_customers

# Display the formatted Monthly Business KPI Report
print("=" * 40)
print("      MONTHLY BUSINESS KPI REPORT")
print("=" * 40)
print()
print(f"Company Name        : {company_name}")
print(f"Reporting Month     : {reporting_month}")
print(f"Prepared By         : {analyst_name}")
print()
print(f"Total Revenue       : ₹{total_revenue:,.2f}")
print(f"Total Cost          : ₹{total_cost:,.2f}")
print(f"Total Profit        : ₹{total_profit:,.2f}")
print(f"Total Orders        : {total_orders}")
print(f"Total Customers     : {total_customers}")
print()
print(f"Profit Margin       : {profit_margin:.2f}%")
print(f"Average Order Value : ₹{average_order_value:,.2f}")
print(f"Revenue Growth      : {revenue_growth:.2f}%")
print(f"Revenue Per Customer: ₹{revenue_per_customer:,.2f}")
print()
print("=" * 40)

# Explain the bonus KPI
print("\nRevenue Per Customer tells management how much revenue is generated on average from each customer,")
print("which helps assess customer value and business performance.")
````
