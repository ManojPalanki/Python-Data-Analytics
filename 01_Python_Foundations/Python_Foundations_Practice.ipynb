## Q7. Calculate Profit Margin

Given:

- Total Sales = `800000`
- Total Cost = `600000`

**Tasks:**

Calculate:

- Total Profit
- Profit Margin

**Formulas:**

Total Profit = Total Sales - Total Cost

Profit Margin = (Total Profit / Total Sales) × 100

**Requirements:**

- Display Total Profit using an f-string.
- Display Profit Margin using an f-string.
- Format Profit Margin to exactly `2` decimal places.

```python
total_sales = 800000
total_cost = 600000
total_profit = total_sales - total_cost
profit_margin = (total_profit / total_sales) * 100
print(f"Total Profit: ₹{total_profit}")
print(f"Profit Margin: {profit_margin:.2f}%")
```

## Q11. Sales KPI Summary

Create the following variables:

- Company Name = `"Tata Retail"`
- Total Sales = `1250000`
- Total Cost = `875000`
- Total Orders = `2500`

Calculate:

- Total Profit
- Profit Margin
- Average Order Value

**Requirements:**

- Use meaningful `snake_case` variable names.
- Use f-strings.
- Display Profit Margin with exactly `2` decimal places.
- Display Average Order Value with exactly `2` decimal places.
- Create a clean Sales KPI Summary.

```python
# Business Data

company_name = "Tata Retail"
total_sales = 1250000
total_cost = 875000
total_orders = 2500

# KPI Calculations

total_profit = total_sales - total_cost
profit_margin = (total_profit / total_sales) * 100
average_order_value = total_sales / total_orders

# Sales KPI Summary

print("\n" + "=" * 50)
print("                 SALES KPI SUMMARY")
print("=" * 50)

print(f"Company:                  {company_name}")
print(f"Total Sales:              ₹{total_sales:,.2f}")
print(f"Total Cost:               ₹{total_cost:,.2f}")
print(f"Total Profit:             ₹{total_profit:,.2f}")
print(f"Profit Margin:            {profit_margin:.2f}%")
print(f"Average Order Value:      ₹{average_order_value:,.2f}")

print("=" * 50)
```

## Q13. Convert Imported Business Data

Assume the following values were imported from a CSV file:

revenue = "1750000"  
cost = "1250000"  
orders = "3500"  
customers = "2100"  

**Tasks:**

Convert all variables to appropriate numerical data types.

Calculate:

- Total Profit
- Profit Margin
- Average Order Value
- Revenue Per Customer

**Requirements:**

- Use meaningful variable names.
- Display every KPI using f-strings.
- Format all calculated decimal values to exactly `2` decimal places.

```python
revenue = "1750000"  
cost = "1250000"  
orders = "3500"  
customers = "2100"  

#Convert all variables to appropriate numerical data types.
revenue = float(revenue)
cost = int(cost)
orders = int(orders)
customers = int(customers)
total_profit = revenue - cost
profit_margin = (total_profit / revenue) * 100
average_order_value = revenue / orders
revenue_per_customer = revenue / customers
print(f"Total Profit: ₹{total_profit}")
print(f"Profit Margin: {profit_margin:.2f}%")
print(f"Average Order Value: ₹{average_order_value:.2f}")
print(f"Revenue per Customer: ₹{revenue_per_customer:.2f}")
```

## Q14. Dynamic Business KPI Report

Ask the user to enter:

- Company Name
- Region
- Total Revenue
- Total Cost
- Total Orders
- Total Customers

Convert all numerical inputs to appropriate numerical data types.

Calculate:

- Total Profit
- Profit Margin
- Average Order Value
- Revenue Per Customer

**Requirements:**

- Use meaningful `snake_case` variable names.
- Use `input()` to accept data.
- Perform appropriate type conversion.
- Add useful comments to separate the Input, KPI Calculation, and Output sections.
- Use f-strings.
- Format calculated decimal values to exactly `2` decimal places.
- Display a clean Business KPI Report.

```python

company_name = input("Enter Company Name: ")
region = input("Enter Region: ")

total_revenue = float(input("Enter Total Revenue: "))
total_cost = float(input("Enter Total Cost: "))
total_orders = int(input("Enter Total Orders: "))
total_customers = int(input("Enter Total Customers: "))


# KPI CALCULATION SECTION

total_profit = total_revenue - total_cost
profit_margin = (total_profit / total_revenue) * 100
average_order_value = total_revenue / total_orders
revenue_per_customer = total_revenue / total_customers

# BUSINESS KPI REPORT

print("\n" + "="*50)
print(f"      BUSINESS KPI REPORT: {company_name.upper()}      ")
print("="*50)
print(f"Region:                  {region}")
print(f"Total Revenue:           ₹{total_revenue:,.2f}")
print(f"Total Cost:              ₹{total_cost:,.2f}")
print(f"Total Orders:            {total_orders:,}")
print(f"Total Customers:         {total_customers:,}")
print("-"*50)
print(f"Total Profit:            ₹{total_profit:,.2f}")
print(f"Profit Margin:           {profit_margin:.2f}%")
print(f"Average Order Value:     ₹{average_order_value:,.2f}")
print(f"Revenue Per Customer:    ₹{revenue_per_customer:,.2f}")
print("="*50)
```

## Q15. Monthly Business Performance Report

### Scenario

You are a Junior Data Analyst working for an e-commerce company.

Your manager gives you the following monthly business data:

- Company Name = `"ShopKart India"`
- Reporting Month = `"June 2026"`
- Total Revenue = `2850000`
- Total Cost = `2075000`
- Total Orders = `4750`
- Total Customers = `3200`

### Your Task

Store all business data using meaningful variables.

Calculate:

- Total Profit
- Profit Margin
- Average Order Value
- Revenue Per Customer

### Requirements

- Use meaningful `snake_case` variable names.
- Use appropriate Python data types.
- Add useful comments.
- Use f-strings for the report.
- Display Profit Margin with exactly `2` decimal places.
- Display Average Order Value with exactly `2` decimal places.
- Display Revenue Per Customer with exactly `2` decimal places.
- Create a clean Monthly Business Performance Report.
- Write the entire solution yourself without copying code from previous questions.

```python
# Business Data

company_name = "ShopKart India"
reporting_month = "June 2026"
total_revenue = 2850000
total_cost = 2075000
total_orders = 4750
total_customers = 3200

# KPI Calculations

total_profit = total_revenue - total_cost
profit_margin = (total_profit / total_revenue) * 100
average_order_value = total_revenue / total_orders
revenue_per_customer = total_revenue / total_customers

# Monthly Business Performance Report

print("\n" + "=" * 60)
print("            MONTHLY BUSINESS PERFORMANCE REPORT")
print("=" * 60)

print(f"Company:                  {company_name}")
print(f"Reporting Month:          {reporting_month}")

print("-" * 60)

print(f"Total Revenue:            ₹{total_revenue:,.2f}")
print(f"Total Cost:               ₹{total_cost:,.2f}")
print(f"Total Profit:             ₹{total_profit:,.2f}")

print("-" * 60)

print(f"Profit Margin:            {profit_margin:.2f}%")
print(f"Total Orders:             {total_orders:,}")
print(f"Total Customers:          {total_customers:,}")
print(f"Average Order Value:      ₹{average_order_value:,.2f}")
print(f"Revenue Per Customer:     ₹{revenue_per_customer:,.2f}")

print("=" * 60)

```
