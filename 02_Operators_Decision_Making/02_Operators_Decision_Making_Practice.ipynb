## Question 8 — Product Inventory Status

Given:

* Current Stock = `45`
* Reorder Level = `50`
* Maximum Stock Capacity = `200`

Write a Python program to classify inventory status:

* Stock equal to `0` → `"Out of Stock"`
* Stock less than or equal to Reorder Level → `"Reorder Required"`
* Stock greater than Maximum Stock Capacity → `"Overstocked"`
* Otherwise → `"Stock Level Normal"`

```python
current_stock = 45
reorder_level = 50
max_stock_capacity = 200

if current_stock == 0:
    print("Out of Stock")
elif current_stock <= reorder_level:
    print("Reorder Required")
elif current_stock > max_stock_capacity:
    print("Overstocked")
else:
    print("Stock Level Normal")
```

## Question 9 — Sales Territory Target Analysis

A sales territory has:

* Sales Target = `₹800,000`
* Actual Sales = `₹920,000`

Write a Python program to:

* Calculate the target achievement percentage.
* Classify performance:

  * Achievement ≥ `120%` → `"Outstanding"`
  * Achievement ≥ `100%` → `"Target Achieved"`
  * Achievement ≥ `80%` → `"Near Target"`
  * Otherwise → `"Below Target"`

```python
sales_target = 800000
actual_sales = 920000

target_achievement = (actual_sales / sales_target) * 100
print(f"Target Achievement: {target_achievement:.2f}%")
if target_achievement >= 120:
    print("Outstanding Performance")    
elif target_achievement >= 100:
    print("Target Achieved")
elif target_achievement >= 80:
    print("Near Target")
else:
    print("Below Target")
```

## Question 10 — Management Attention Flag

A company wants to automatically identify products requiring management attention.

Given:

* Product Revenue = `₹200,000`
* Product Cost = `₹230,000`
* Units Sold = `40`
* Inventory = `300`

A product requires management attention when:

* The product is making a loss.

**OR**

* Units sold are less than `50` and inventory is greater than `200`.

Write a Python program to:

1. Calculate product profit.
2. Check both business conditions.
3. Print `"Management Attention Required"` when either condition is true.
4. Otherwise, print `"Product Performance Normal"`.

```python
product_revenue = 200000
product_cost = 230000
units_sold = 40
product_inventory = 300

product_profit = product_revenue - product_cost

is_loss = product_profit < 0
low_sales_high_inventory = units_sold < 50 and product_inventory > 200

if is_loss or low_sales_high_inventory:
    print("Management Attention Required")
else:
    print("Product Performance Normal")
```
