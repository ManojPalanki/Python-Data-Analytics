````python
product_name = "Wireless Headphones"
units_sold = 850
selling_price = 2500
cost_price = 1800
monthly_sales_target = 2000000
current_inventory = 300
reorder_level = 100
customer_rating = 4.2

total_revenue = units_sold * selling_price
total_cost = units_sold * cost_price
total_profit = total_revenue - total_cost

profit_margin = (total_profit / total_revenue) * 100 if total_revenue != 0 else 0
target_achievement = (total_revenue / monthly_sales_target) * 100 if monthly_sales_target != 0 else 0

if profit_margin >= 25:
    profit_performance = "High Profitability"
elif profit_margin >= 15:
    profit_performance = "Moderate Profitability"
elif profit_margin > 0:
    profit_performance = "Low Profitability"
else:
    profit_performance = "Loss-Making Product"

if target_achievement >= 120:
    sales_performance = "Outstanding Performance"
elif target_achievement >= 100:
    sales_performance = "Target Achieved"
elif target_achievement >= 80:
    sales_performance = "Near Target"
else:
    sales_performance = "Below Target"

if current_inventory == 0:
    inventory_status = "Out of Stock"
elif current_inventory <= reorder_level:
    inventory_status = "Reorder Required"
elif current_inventory > 500:
    inventory_status = "Overstocked"
else:
    inventory_status = "Inventory Level Normal"

if customer_rating >= 4.5:
    customer_status = "Excellent Customer Satisfaction"
elif customer_rating >= 4.0:
    customer_status = "Good Customer Satisfaction"
elif customer_rating >= 3.0:
    customer_status = "Average Customer Satisfaction"
else:
    customer_status = "Poor Customer Satisfaction"

management_alert = (
    "Management Attention Required"
    if total_profit < 0
    or target_achievement < 80
    or customer_rating < 3.0
    or current_inventory > 500
    else "Product Performance Normal"
)

if profit_margin >= 25 and target_achievement >= 100 and customer_rating >= 4.0:
    overall_performance = "Excellent Product"
elif profit_margin >= 15 and target_achievement >= 80 and customer_rating >= 3.5:
    overall_performance = "Good Product"
else:
    overall_performance = "Needs Improvement"

print("========================================")
print("       PRODUCT PERFORMANCE REPORT")
print("========================================")
print()
print(f"Product Name        : {product_name}")
print()
print(f"Total Revenue       : ₹{total_revenue:,}")
print(f"Total Cost          : ₹{total_cost:,}")
print(f"Total Profit        : ₹{total_profit:,}")
print()
print(f"Profit Margin       : {profit_margin:.2f}%")
print(f"Target Achievement  : {target_achievement:.2f}%")
print()
print(f"Profit Performance  : {profit_performance}")
print(f"Sales Performance   : {sales_performance}")
print(f"Inventory Status    : {inventory_status}")
print(f"Customer Rating     : {customer_status}")
print()
print(f"Management Alert    : {management_alert}")
print()
print(f"Overall Performance : {overall_performance}")
print()
print("========================================")
````

````python
# Business case variables (reuse existing notebook variables where appropriate)
company_name = "Nova Retail Pvt. Ltd."
reporting_month = "June 2026"

monthly_revenue = total_revenue        # from previous cell
monthly_cost = total_cost              # from previous cell
sales_target = monthly_sales_target    # from previous cell
total_orders = units_sold              # reuse units_sold as total orders
customer_satisfaction = customer_rating
product_category = "Electronics"

# KPI calculations
profit = monthly_revenue - monthly_cost
profit_margin_pct = (profit / monthly_revenue) * 100 if monthly_revenue != 0 else 0
average_order_value = monthly_revenue / total_orders if total_orders != 0 else 0

# Sales target status
target_status = "Target Achieved" if monthly_revenue >= sales_target else "Target Missed"

# Profitability classification
if profit_margin_pct >= 30:
    profitability = "High Profit"
elif profit_margin_pct >= 15:
    profitability = "Medium Profit"
elif profit_margin_pct >= 0:
    profitability = "Low Profit"
else:
    profitability = "Loss"

# Business recommendation
if target_status == "Target Achieved" and profitability == "High Profit":
    recommendation = "Expand Product"
elif target_status == "Target Achieved" and profitability == "Medium Profit":
    recommendation = "Maintain Performance"
elif target_status == "Target Missed" and profit > 0:
    recommendation = "Improve Marketing"
elif profitability == "Loss":
    recommendation = "Discontinue Product"
else:
    recommendation = "Monitor Performance"

# Final report
print("=========================================")
print("        NOVA RETAIL PERFORMANCE REPORT")
print("=========================================")
print()
print(f"Company                 : {company_name}")
print(f"Product                 : {product_name}")
print(f"Reporting Month         : {reporting_month}")
print()
print(f"Revenue                 : ₹{monthly_revenue:,}")
print(f"Cost                    : ₹{monthly_cost:,}")
print(f"Profit                  : ₹{profit:,}")
print()
print(f"Profit Margin           : {profit_margin_pct:.2f}%")
print()
print(f"Total Orders            : {total_orders}")
print(f"Average Order Value     : ₹{average_order_value:,.2f}")
print()
print(f"Sales Target            : ₹{sales_target:,}")
print(f"Target Status           : {target_status}")
print()
print(f"Customer Satisfaction   : {customer_satisfaction} / 5")
print()
print(f"Product Category        : {product_category}")
print()
print(f"Profitability           : {profitability}")
print()
print(f"Business Recommendation : {recommendation}")
print()
print("=========================================")
````
