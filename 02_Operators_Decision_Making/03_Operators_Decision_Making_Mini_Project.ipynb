# Mini Project — Retail Product Performance & Management Alert System

## Project Overview

A retail company wants to automatically evaluate product performance using sales, revenue, cost, inventory, and customer rating data.

Currently, managers manually review product performance. Your task is to build a Python program that analyzes a product and generates a performance summary using Python operators and decision-making statements.

---

## Skills Covered

* Variables
* Data Types
* Arithmetic Operators
* Comparison Operators
* Logical Operators
* Assignment Operators
* `if`
* `elif`
* `else`
* Nested Conditions
* Business Logic
* KPI Calculations

---

## Business Scenario

The management team wants to evaluate the performance of a product.

Given the following product information:

* Product Name = `Wireless Headphones`
* Units Sold = `850`
* Selling Price per Unit = `₹2,500`
* Cost per Unit = `₹1,800`
* Monthly Sales Target = `₹2,000,000`
* Current Inventory = `300`
* Reorder Level = `100`
* Customer Rating = `4.2`

Build a Python program that analyzes the product and generates a management performance report.

# Tasks

## Task 1 — Store Product Data

Create variables to store:

* Product Name
* Units Sold
* Selling Price per Unit
* Cost per Unit
* Monthly Sales Target
* Current Inventory
* Reorder Level
* Customer Rating

Use appropriate Python data types.

---

## Task 2 — Calculate Business KPIs

Calculate:

* Total Revenue
* Total Cost
* Total Profit
* Profit Margin Percentage
* Sales Target Achievement Percentage

Use arithmetic operators.

### Formulas

`Total Revenue = Units Sold × Selling Price`

`Total Cost = Units Sold × Cost Price`

`Total Profit = Total Revenue - Total Cost`

`Profit Margin = (Total Profit / Total Revenue) × 100`

`Target Achievement = (Total Revenue / Sales Target) × 100`

---

## Task 3 — Classify Profit Performance

Use `if`, `elif`, and `else`.

Classification rules:

* Profit Margin ≥ `25%` → `"High Profitability"`
* Profit Margin ≥ `15%` → `"Moderate Profitability"`
* Profit Margin > `0%` → `"Low Profitability"`
* Otherwise → `"Loss-Making Product"`

---

## Task 4 — Analyze Sales Target Performance

Classification rules:

* Target Achievement ≥ `120%` → `"Outstanding Performance"`
* Target Achievement ≥ `100%` → `"Target Achieved"`
* Target Achievement ≥ `80%` → `"Near Target"`
* Otherwise → `"Below Target"`

---

## Task 5 — Analyze Inventory Status

Classification rules:

* Inventory = `0` → `"Out of Stock"`
* Inventory ≤ Reorder Level → `"Reorder Required"`
* Inventory > `500` → `"Overstocked"`
* Otherwise → `"Inventory Level Normal"`

---

## Task 6 — Analyze Customer Rating

Classification rules:

* Rating ≥ `4.5` → `"Excellent Customer Satisfaction"`
* Rating ≥ `4.0` → `"Good Customer Satisfaction"`
* Rating ≥ `3.0` → `"Average Customer Satisfaction"`
* Otherwise → `"Poor Customer Satisfaction"`

---

## Task 7 — Create Management Attention Flag

The product requires management attention when:

* The product is making a loss.

**OR**

* Target Achievement is below `80%`.

**OR**

* Customer Rating is below `3.0`.

**OR**

* Inventory is greater than `500`.

Create a condition using logical operators.

Output:

`"Management Attention Required"`

or

`"Product Performance Normal"`

---

## Task 8 — Determine Overall Product Performance

Classify the overall product performance.

### Excellent Product

The product must satisfy all conditions:

* Profit Margin ≥ `25%`
* Target Achievement ≥ `100%`
* Customer Rating ≥ `4.0`

### Good Product

The product must satisfy:

* Profit Margin ≥ `15%`
* Target Achievement ≥ `80%`
* Customer Rating ≥ `3.5`

### Needs Improvement

If neither condition is satisfied.

Use logical operators and decision-making statements.
